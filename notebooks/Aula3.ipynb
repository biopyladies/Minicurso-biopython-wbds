**<font color='red'>O que é o Entrez?:</font>**

Sintaxe: ``bio.Entrez.Utilitario``

- É um sistema de **<font color='red'>recuperação de dados</font>** que fornece aos usuários acesso aos bancos de dados do NCBI, como **<font color='red'>PubMed, GenBank, GEO, PDB e outros</font>**; 
- Formas de acesso:
  - Via navegador da web para inserir consultas manualmente; 
  - Usando o módulo Entrez do pacote Biopython **<font color='red'>Bio.Entrez do Biopython</font>**. Aoresenta vários módulos que permitem pesquisar no PubMed ou baixar os registros do GenBank a partir de um script Python.

**<font color='red'>E como funciona?:</font>**
- Usa os utilitários de programação Entrez (também conhecidos como **<font color='red'>EUtils</font>**), que consistem em várias ferramentas que são descritas em detalhes na página do NCBI: https://www.ncbi.nlm.nih.gov/books/NBK25501/
- Este módulo garante que uma URL correta ("Uniform Resource Locator" ou "Localizador Uniforme de Recursos") seja usada para consultas e que as diretrizes do NCBI para o acesso responsável aos dados sejam seguidas.

**<font color='red'>E o que são essas utilidades do EUtils?:</font>**
- Cada uma dessas 8 ferramentas corresponde a uma função Python do módulo Bio.Entrez;
- Formato de saída: **<font color='red'>XML</font>**;

**<font color='red'>E o que é o XML (Extensible Markup Language)?:</font>**
- É uma linguagem de marcação que define um conjunto de regras para a codificação de documentos em um formato tanto legível por humanos quanto por máquinas.

**<font color='red'>Como analiso um XML?:</font>**
- Usando o Bio.Entrez --> Seguro e eficaz;
- Usando um dos analisadores XML disponíveis na biblioteca padrão do Python (https://docs.python.org/3/library/xml.html) --> Apresenta problemas de segurança --> Cuidado! Os módulos XML não são seguros contra dados errôneos ou construídos de forma maliciosa;
- Lendo a saída XML como texto bruto e analisando-o por pesquisa e manipulação de strings --> Perda de tempo.

**<font color='red'>Como abro um arquivo XML?:</font>**
- Os arquivos XML (formato de texto) devem ser abertos em modo binário para que o analisador XML em Bio.Entrez funcione corretamente:

``open ("myfile.xml", "rb"))``

**<font color='red'>E o que são esses formatos de texto e binários?:</font>**

Entendendo a diferença entre **<font color='red'>arquivo de texto</font>** e **<font color='red'>arquivo binário:</font>**

**<font color='red'>Arquivos de texto:</font>**
- Os dados são representados linha à linha;
- É uma sequência de bytes representando caracteres; 
- As linhas são representadas pelo caracter de quebra de linha, por exemplo: '\n' que varia de sistema operacional para sistema operacional;
- Todos os dados são armazenados como caracteres. por exemplo o número 42 utilizaria um byte para cada dígito do número.

**<font color='red'>Arquivos binários:</font>**
- São representados por uma sequência de bytes sem o conceito de quebra de linha;
- Armazena o dado literal (não em forma de caracteres); 
- Ex.: O número 42 ocupa apenas 1 byte cujo valor em binário seria 101010.

**<font color='red'>Como processar arquivos de texto?:</font>**

- Necessitam de **parsing** (análise caractere à caractere) para identificar as quebras de linha;
- Não é necessário conhecimento prévio sobre a disposição dos dados;

**<font color='red'>Como processar arquivos binários?:</font>**
- É prescindível ter conhecimento sobre os dados, uma vez que o processamento é feito por **deslocamento**;
- Exemplo: É necessário saber que os 2 primeiros bytes representam um número Inteiro. Caso contrário é apenas um conjunto de bytes sequenciais e sem significado. 

**<font color='red'>Outras características:</font>**
- Os arquivos de texto são mais legíveis tanto por humanos quanto por máquinas;
- Os arquivos binários fornecem maior desempenho ao serem lidos por máquinas;
- Os arquivos de texto são mais facilmente depuráveis quando há erros;
- Os arquivos binários tendem a ocupar maior espaço de memória;
- Exemplos:
   - **Formatos de texto:**
    - Linguagens de marcação: HTML, Markdown;
    - Formatos de mensagens: JSON, XML;
    - Formatos de dados: csv, yaml, txt

   - **Formatos binários:**
     - Codificadores de vídeo: Xvid, x264, MPEG;
     - Codificadores de áudio: FLAC, LAME;
     - Compressores de arquivos: gzip, 7z, xz;
     - Codificadores de imagem: PNG, JPEG, GIF, WebP;
     - Serviços: HTTP 2, gRPC.

**<font color='red'>E o que são os arquivos DTDs?:</font>**

- O NCBI usa arquivos **<font color='red'>DTD (Document Type Definition)</font>** para descrever a estrutura das informações contidas nos arquivos XML;
- Um arquivo DTD é um arquivo de texto que especifica elementos e atributos usados em um documento XML, SGML, HTML ou XHTML;
- Ele também pode definir a ordem e o aninhamento de elementos e seus atributos;
- Os arquivos DTD são comumente referenciados em páginas da web por meio de uma URL;
- A maioria dos arquivos DTD usados ​​pelo NCBI estão incluídos na distribuição Biopython;
- O analisador Bio.Entrez usa os arquivos DTD (arquivo de texto) ao analisar um arquivo XML (arquivo de texto) retornado pelo NCBI Entrez mas lembrando que os arquivos devem ser abertos em modo **binário** para que o analisador XML em Bio.Entrez funcione corretamente:

``open ("myfile.xml", "rb")) # Converte os arquivos de texto em binários``

- Ocasionalmente, algum arquivo DTD associado a um arquivo XML específico pode estar ausente na distribuição Biopython. Isso acontece quando o NCBI atualiza seus arquivos DTD e Entrez.read mostrará uma mensagem de aviso com o nome e a URL do arquivo DTD ausente. 

**<font color='red'>Resumindo:</font>**
- Os utilitários de programação Entrez também podem gerar saída em outros formatos, como **Fasta** ou **GenBank** para **bancos de dados de sequências** ou o formato **MedLine** para **bancos de dados de literatura**;
- Portanto, o módulo Bio.Entrez:
  - Aceita dados em **<font color='red'>formato binário</font>**; 
  - **<font color='red'>Extrai a codificação do XML</font>** e a usa para decodificar todo o texto no documento XML para strings Python, garantindo textos em idiomas diferentes do inglês sejam interpretados corretamente.

**<font color='red'>Pubmed:</font>** **<font color='blue'>Banco de dados de citações e resumos da literatura biomédica da MEDLINE e de outras revistas científicas da área biológica. **<font color='red'>Protein:</font>** O banco de dados de proteínas contém sequências de aminoácidos criadas a partir das traduções de regiões codificantes fornecidas nos registros de nucleotídeos no GenBank, EMBL e DDBJ, os membros da Colaboração de Bancos de Dados de Sequência Internacional de Nucleotídeos (INSDC), regiões codificantes em Sequências de Referência do NCBI e os registros do banco de dados de anotação de terceiros (TPA). **<font color='red'>Structure:</font>** Lista todas as sequências de estruturas proteicas resolvidas experimentalmente por cristalográfica ou RMN depositados no PDB;

**<font color='red'>Genome:</font>** Contém dados de sequências e mapas de todos os genomas completamente sequenciados e com sequenciamento em andamento de mais de 1000 espécies ou cepas. Os três domínios principais da vida (bactérias, arquéias e eucariotos) estão representados, bem como muitos vírus, fagos, viróides, plasmídeos e organelas;

Os bancos de dados de origem para domínios conservados são Pfam, Smart e COG;

**<font color='red'>gap:</font>** Fornece os resultados de estudos que investigaram a interação entre o genótipo e o fenótipo, incluindo estudos de associação do genoma, sequenciamento médico, ensaios de diagnóstico molecular e associação entre genótipos com características não clínicas;

**<font color='red'>gene:</font>** Lista de genomas que foram completamente sequenciados e que têm uma comunidade de pesquisa ativa para contribuir com dados específicos de genes. As informações incluem nomenclatura, localização cromossômica, produtos de genes e seus atributos (por exemplo, interações de proteínas), marcadores associados, fenótipos, interações e links para citações, sequências, detalhes de variação, mapas, relatórios de expressão, homólogos, conteúdo de domínio de proteína, e bancos de dados externos;

**<font color='red'>gds e geo:</font>** Existem 2 bancos GEO: GEO Database e GEO Profiles. Armazenam conjuntos de dados selecionados de expressão gênica e abundância molecular de dados obtidos por microarray montados pelo NCBI a partir do repositório Gene Expression Omnibus (GEO);

**<font color='red'>homologene:</font>** Contém conjuntos gerados automaticamente de genes homólogos e seus respectivos mRNA, dados genômicos e de sequência de proteínas de organismos eucarióticos selecionados. Potenciais homólogos de outros organismos são incluídos por meio de similaridade de sequência com clusters do UniGene;

**<font color='red'>mesh:</font>** MeSH (Medical Subject Headings) é o vocabulário controlado da National Library of Medicine e o sistema de classificação (ontologia) usado para indexar artigos no PubMed e para obter informações que podem usar terminologia diferente para os mesmos conceitos;

**<font color='red'>ncbisearch:</font>** NCBI Site Search é um banco de dados de páginas da web estáticas do NCBI, documentação e ferramentas online para pesquisar por ferramentas especializadas em análise de sequências, edições anteriores de boletins informativos, páginas de descrição de recursos legados, código de amostra e outros;

**<font color='red'>omim:</font>** O banco de dados OMIM (Online Mendelian Inheritance in Man) permite pesquisas de artigos OMIM sobre genes humanos, doenças genéticas e outras características herdadas. Os artigos do OMIM fornecem links para referências de literatura associadas, registros de sequência, mapas e bancos de dados relacionados. Os registros do OMIM são hospedados e disponibilizados pelo site independente do OMIM (www.omim.org);

**<font color='red'>pmc:</font>** O PubMed Central (PMC) é o arquivo digital de literatura de periódicos de ciências biológicas da Biblioteca Nacional de Medicina dos EUA. PMC contém manuscritos de textos completos depositados por autores ou artigos fornecidos pela editora;

**<font color='red'>popset:</font>** O banco de dados PopSet contém sequências de nucleotídeos relacionadas que se originam de estudos comparativos: filogenia populacional, ambiental e mutacional. Cada registro no banco de dados é um conjunto de sequências de nucleotídeos representando a mesma molécula da mesma espécie (população, mutação) em diferentes espécies identificáveis (filogenética) ou espécies anônimas da mesma comunidade biológica (ecossistema);

**<font color='red'>probe:</font>** Banco de dados de reagentes de ácidos nucleicos projetado para ser usado em uma ampla variedade de aplicações de pesquisa biomédica, incluindo genotipagem, estudos de expressão gênica, descoberta de SNPs, mapeamento de genoma e silenciamento de genes com informações sobre distribuidores de reagentes, eficácia da sonda e semelhanças de sequência computadas.

**<font color='red'>proteinclusters:</font>** Coleção de sequências de proteínas relacionadas (clusters) que consistem em proteínas de sequência de referência que são codificadas por genomas procarióticos completos, bem como por plasmídeos e genomas de organelas eucarióticas codificados. A taxonomia fornece o vocabulário controlado principalmente para classificar dados moleculares no sistema Entrez;

**<font color='red'>unigene:</font>** Fornece conjuntos não redundantes gerados automaticamente (clusters) de sequências de transcrição, cada cluster representando um locus de transcrição distinto (gene ou pseudogene expresso). ## **<font color='red'>Utilitário 2: ESearch: Buscando por valores nos db</font>**

Para pesquisar qualquer um desses bancos de dados, usamos `Bio.Entrez.esearch()`. 

Por exemplo, vamos pesquisar no PubMed por publicações que incluem Biopython em seu título:

````python
# Pesquisando por publicações depositadas no PubMed que apresentam a palavra "Biopython" no título...
handle = Entrez.esearch(db="pubmed", term="biopython[title]", retmax="40")
record = Entrez.read(handle)

# Observando os IDs do PubMed com Biopython...
print(record["IdList"]) 
````

Fazendo uma pesquisa pelo gene matK em orquídeas *Cypripedioideae*

````python
# Contabilizando o número de vezes que aparece o gene matK de Cypripedioideae no db "nucleotide"...
handle = Entrez.esearch(db="nucleotide", term="Cypripedioideae[Orgn] AND matK[Gene]")
record = Entrez.read(handle)
record["Count"]
````

````python
# Observando cada ID do GenBank correspondente à busca anterior...
record["IdList"]
````

Observe que, em vez de um nome de espécie como `Cypripedioideae[Orgn]`, podemos restringir a pesquisa usando um identificador de táxon NCBI, aqui seria `txid158330[Orgn]`. 

Incluir `complete[prop]` em uma pesquisa de genoma restringe-se apenas a genomas completos.

Como um exemplo final, vamos obter uma lista de títulos de periódicos computacionais:

````python
# Obtendo o valor de todas as revistas na área computacional depositadas no db "nlmcatalog"...
handle = Entrez.esearch(db="nlmcatalog", term="computational[Journal]")
record = Entrez.read(handle)
print("{} revistas computacionais encontradas".format(record["Count"]))
````

````python
# Obtendo os IDs das primeiras 20 revistas...
print("As primeiras 20 revistas têm como ID:\n{}".format(record["IdList"]))
````

## **<font color='red'>Utilitário 3: EPost: Manipulando listas longas de requerimento</font>**

- EPost carrega uma lista grande de IDs, disponível por meio da função `Bio.Entrez.epost()`;
1. Suponha que tenha uma lista bem longa de IDs que deseja baixar (ex.: sequências ou citações);
2. Quando você faz uma solicitação com EFetch, sua lista de IDs, o banco de dados, etc, são todos transformados em uma longa URL enviada ao servidor. Se sua lista de IDs for longa, este URL ficará longo, e URLs longos podem quebrar (por exemplo, alguns prefixos não funcionam bem);
3. Em vez disso, você pode dividir isso em duas etapas:
  - Fazendo o upload da lista de IDs usando EPost (usa um “post HTML” internamente, em vez de um “get HTML”, contornando o problema da URL longa);
  - Com o suporte de histórico, você pode consultar esta lista de IDs e baixar os dados associados com EFetch.

## **<font color='red'>Utilitários 10 e 11: Entrez.read e Entrez.parse: Caracteres de escape HTML e Manipulando erros</font>**


- Os registros Pubmed podem conter tags HTML para indicar, por exemplo, subscritos, sobrescritos ou texto em itálico ou símbolos matemáticos via MathML;
- Por padrão, o analisador `Bio.Entrez()` trata todo o texto como simples sem marcação; por exemplo, o fragmento:

`“P <0,05”` 

no resumo de um registro Pubmed, que é codificado como:

`<i> P </i> & lt; 0,05`

no XML retornado por Entrez, é convertido na string Python:

`'<i> P </i> <0,05'`

Embora seja mais legível por humanos, não é um HTML válido devido ao sinal <

Para garantir que todas as strings retornadas pelo analisador sejam HTML válidos, deve ser usada a função `Entrez.read()` ou `Entrez.parse()` com o argumento de escape definido como True:

```
registro = Entrez.read(handle, validate=True, escape=False)
```

O analisador irá então substituir todos os caracteres não permitidos em HTML por seus equivalentes;
- No exemplo acima, o analisador irá gerar:

`'<i> P </i> & lt; 0,05 '`

que é um fragmento de HTML válido. 

Por padrão, o escape é False.

3 coisas podem dar errado ao analisar um arquivo XML:
- O arquivo pode não ser um arquivo XML para começar;
- O arquivo pode terminar prematuramente ou ser corrompido de outra forma;
- O arquivo pode ser um XML correto, mas contém itens que não são representados no DTD associado.
 
**Erro do tipo I (O arquivo pode não ser um arquivo XML):**

O primeiro caso ocorre se tentar analisar um arquivo Fasta como se fosse um arquivo XML:

````python
# Tratando um arquivo Fasta como XML...
handle = open("NC_005816.fna", "rb") # um arquivo fasta
record = Entrez.read(handle)
````

Assim, o analisador não encontrou a tag `<?Xml...` com a qual um arquivo XML deve iniciar e mostra corretamente que o arquivo não é um arquivo XML.

**Erro do tipo II (Arquivos corrompidos):**

Quando o arquivo está no formato XML, mas está corrompido (por exemplo, terminando prematuramente), o analisador irá gerar um `CorruptedXMLError`. 

Segue um exemplo de um arquivo XML que termina prematuramente:

Observe que a mensagem de erro informa em que ponto do arquivo XML o erro foi detectado.

**Erro do tipo III (Com tags no arquivo XML e sem descrição no arquivo DTD):**

O terceiro tipo de erro ocorre quando o arquivo XML contém tags que não têm uma descrição no arquivo DTD correspondente.

## **<font color='red'>Forma de saída: Bio.Medline: Analisando registros Medline</font>**


No seguinte exemplo, consultaremos o PubMed para todos os artigos relacionados com orquídeas. 

Primeiro verificamos quantos desses artigos existem:

````python
# Observando a quantidade de artigos relacionados ao termo "orchid"...
handle = Entrez.egquery(term="orchid")
record = Entrez.read(handle)
for row in record["eGQueryResult"]:
  if row["DbName"]=="pubmed":
    print(row["Count"])
````

Agora vamos usar a função `Bio.Entrez.efetch()` para baixar os IDs PubMed desses 2663 artigos:

````python
# Baixando os IDs desses artigos...
handle = Entrez.esearch(db="pubmed", term="orchid", retmax=463)
record = Entrez.read(handle)
handle.close()
idlist = record["IdList"]
print(idlist)
````

Agora que temos os IDs, vamos obter os registros Medline correspondentes e extrair as informações deles. 

Primeiro, vamos baixar os registros do Medline no formato de arquivo simples do Medline e usar o módulo `Bio.Medline()` para analisá-los:

````python
# Baixando o Medline...
from Bio import Medline
````

````python
# Criando as variáveis...
handle = Entrez.efetch(db="pubmed", id=idlist, rettype="medline", retmode="text")
records = Medline.parse(handle)
````

Obs.: Se quiser salvar os registros, você pode convertê-los em uma lista:


````python
records = list(records)
````

Vamos agora iterar os registros para imprimir algumas informações sobre cada um:

````python
for record in records:
  print("title:", record.get("TI", "?"))
  print("authors:", record.get("AU", "?"))
  print("source:", record.get("SO", "?"))
  print("")
````
