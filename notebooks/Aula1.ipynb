A palavra-chave "self" nos permite acessar os atributos e m√©todos da classe em Python. "self" relaciona a classe a seu objeto.

````python
class Transgenico: 
  def __init__(self, organismo, dna_organismo):
    self.organismo = organismo
    self.dna_organismo = dna_organismo
    
milho = Transgenico("milho", "ACATCTGATCAGCTACGATCGATCAGCTAGCAT")
print(milho)
````

A vari√°vel milho se tornou um **objeto da classe Transg√™nico**, para o Python isso faz todo sentido, mas do ponto de vista biol√≥gico n√£o, pois o que define um transg√™nico √© a adi√ß√£o de parte do DNA de outro organismo. Vamos resolver essa quest√£o com a cria√ß√£o de um **m√©todo** que faz a adi√ß√£o desse DNA.

````python
class Transgenico: #Primeira letra maiuscula
  def __init__(self, organismo, dna_organismo):
    self.organismo = organismo
    self.dna_organismo = dna_organismo
  
  def alterar_dna(self, outro_dna):
    novo_dna = self.dna_organismo + outro_dna
    tamanho = len(novo_dna)
    print(f"O novo dna do organismo √©: {novo_dna}")
    
milho = Transgenico("milho", "ACATCTGATCAGCTACGATCGATCAGCTAGCAT")
print(milho.dna_organismo)
#novo_dna = milho.alterar_dna("AAAAAAAAAAAAAAACT")
#print(milho)
````

## **ORF vs. CDS**

**IMPORTANTE:**
- Uma **ORF (*Open Reading Frame*)** √© uma sequ√™ncia de DNA que come√ßa com o c√≥don de in√≠cio ‚ÄúATG‚Äù (nem sempre) e termina com qualquer um dos tr√™s c√≥dons de termina√ß√£o (TAA, TAG, TGA) (nem sempre);
- Uma defini√ß√£o alternativa diz que uma ORF √© uma sequ√™ncia que tem um comprimento divis√≠vel por tr√™s e √© limitada por c√≥dons de parada;
- A ORF √© prevista geralmente com base na sequ√™ncia de DNA sem ter sido necessariamente comprovada a transcri√ß√£o, ou seja, podemos encontrar ORFs no DNA sem comprovar a exist√™ncia da sua transcri√ß√£o e posterior tradu√ß√£o;
- Uma ORF pode:
  - N√£o ser transcrita (ex.: uma ORF truncada);
  - Ser transcrita e n√£o ser traduzida (ex.: miRNA);
  - Ser transcrita (ex.: mRNA ap√≥s capeamento e adi√ß√£o da cauda-poliA) e traduzida (prote√≠na);
- Dependendo do ponto de partida, existem 6 maneiras poss√≠veis de traduzir qualquer sequ√™ncia de nucleot√≠deos em uma sequ√™ncia de amino√°cidos de acordo com o c√≥digo gen√©tico (3 na fita parental e 3 na complementar), chamadas de **frames de leitura**.
- Uma **CDS (*Coding Sequence*)** √© a regi√£o real do DNA que √© traduzida para formar prote√≠nas (sem os √≠ntros, no caso de eucariotos), em outras palavras, uma CDS refere-se √†queles nucleot√≠deos (exons concatenados) que podem ser divididos em c√≥dons que s√£o traduzidos em amino√°cidos pelo mecanismo de tradu√ß√£o ribossomal.

**Resumindo:**
- Todo CDS deriva de uma ORF, por√©m, n√£o toda ORF gera uma CDS.

## **Start-codon**

**IMPORTANTE:**
- O c√≥don de in√≠cio √© o primeiro c√≥don de um transcrito de RNA mensageiro (mRNA) traduzido por um ribossomo;
- O c√≥don inicial sempre codifica para Metionina (Met) em eucariotos e Archaea e uma N-formilmetionina (fMet) em bact√©rias, mitoc√¥ndrias e plast√≠deos;
- O c√≥don de in√≠cio mais comum √© AUG (isto √©, ATG na sequ√™ncia de DNA correspondente);
- Os c√≥dons de in√≠cio alternativos s√£o diferentes do c√≥don AUG padr√£o e s√£o encontrados tanto em procariotos (bact√©rias e arqu√©ias) quanto em eucariotos;
